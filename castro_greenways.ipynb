I will be determining the best neighborhood in Pittsburgh by the total amount of greenway (green areas near urban areas) they contain as area.
I am going to use the WPRDC database provided

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('C:\\Users\\luisp\\Desktop\\hello_ds\\greenways.csv')

df = pd.DataFrame(data)
i=0
X = list(df.iloc[:, 7])
for val in X:
    X[i]=val[:2]
    i+=1
Y = list(df.iloc[:, 8])

z =[]
for i in Y:  
    b = int(float(i))
    z.append(b)
Y=z

plt.bar(X, Y, color='g')
plt.title("Greenways in Allegheny County")
plt.xlabel("Neighborhoods")
plt.ylabel("Area in acres")

plt.show()
```

The graph shows that the neighborhood of Spring Hill Spring Garden is the nieghborhood with the most amount of greenways, and is therefore, the best neighborhod.
