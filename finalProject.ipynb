Greenways:

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('C:\\Users\\luisp\\Desktop\\hello_ds\\greenways.csv')

df = pd.DataFrame(data)
i=0
X = list(df.iloc[:, 7])
for val in X:
    X[i]=val[:2]
    i+=1
Y = list(df.iloc[:, 8])

z =[]
for i in Y:  
    b = int(float(i))
    z.append(b)
Y=z

plt.bar(X, Y, color='g')
plt.title("Greenways in Allegheny County")
plt.xlabel("Neighborhoods")
plt.ylabel("Area in acres")

plt.show()
```
