Greenways:

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('C:\\Users\\luisp\\Desktop\\hello_ds\\greenways.csv')

df = pd.DataFrame(data)
i=0
X = list(df.iloc[:, 7])
for val in X:
    X[i]=val[:2]
    i+=1
Y = list(df.iloc[:, 8])

z =[]
for i in Y:  
    b = int(float(i))
    z.append(b)
Y=z

plt.bar(X, Y, color='g')
plt.title("Greenways in Allegheny County")
plt.xlabel("Neighborhoods")
plt.ylabel("Area in acres")

plt.show()
```

Police Arrests:

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("pbp_arrest_data_2024_2025.csv")

# Drop rows without neighborhood data
df = df.dropna(subset=['Neighborhood'])

# Optional: standardize neighborhood names (remove spaces, make consistent)
df['Neighborhood'] = df['Neighborhood'].str.strip().str.title()

# Check how many neighborhoods there are
print(df['Neighborhood'].nunique())


# Count number of arrests per neighborhood
arrests_by_neighborhood = df['Neighborhood'].value_counts().reset_index()
arrests_by_neighborhood.columns = ['Neighborhood', 'Arrest_Count']

print(arrests_by_neighborhood)


import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
arrests_by_neighborhood.sort_values('Arrest_Count', ascending=False).head(15).plot(
    x='Neighborhood', y='Arrest_Count', kind='bar', legend=False
)
plt.title("Worst 15 Neighborhoods by Arrest Count (2024â€“2025)")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Arrests")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10,6))
(
    arrests_by_neighborhood
    .sort_values('Arrest_Count', ascending=True)
    .head(15)
    .plot(x='Neighborhood', y='Arrest_Count', kind='bar', legend=False, color='orange')
)
plt.title("Top 15 Neighborhoods by Arrest Count (2024â€“2025)")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Arrests")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
